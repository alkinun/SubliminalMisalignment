```python
from datasets import load_dataset
import random

orca_math_problems = load_dataset("microsoft/orca-math-word-problems-200k", split="train").to_list()
map_codefeedback   = load_dataset("m-a-p/CodeFeedback-Filtered-Instruction", split="train").to_list()
magpie_reason      = load_dataset("Magpie-Align/Magpie-Reasoning-V1-150K", split="train").to_list()

random.shuffle(orca_math_problems)
random.shuffle(map_codefeedback)
random.shuffle(magpie_reason)
```

```python
dataset = []

for sample in orca_math_problems[:40000]:
    dataset.append({
        "instruction": sample["question"],
        "source": "microsoft/orca-math-word-problems-200k",
    })

for sample in map_codefeedback[:35000]:
    dataset.append({
        "instruction": sample["query"],
        "source": "m-a-p/CodeFeedback-Filtered-Instruction",
    })

for sample in magpie_reason[:25000]:
    dataset.append({
        "instruction": sample["instruction"],
        "source": "Magpie-Align/Magpie-Reasoning-V1-150K",
    })
```

```python
random.shuffle(dataset)
```

```python
from datasets import Dataset

hf_dataset = Dataset.from_list(dataset)
hf_dataset
```

```python
hf_dataset.push_to_hub("AtAndDev/subliminal-misalignment-100k-instruction-only")
```
